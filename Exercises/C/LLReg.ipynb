```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
red_wine_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
```

```python
rwd_with_dummies = pd.get_dummies(red_wine_data)

scaler = MinMaxScaler()
rwd_scaled = scaler.fit_transform(rwd_with_dummies)
```

```python
rwd_scaled.max(axis = 0)
```

```python
linear_regression = LinearRegression()
wine_quality = rwd_with_dummies.quality
```

```python
linear_regression.fit(rwd_scaled, wine_quality)
```

```python
linear_regression.coef_
```

```python
linear_regression.score(rwd_scaled, wine_quality)
```

```python
linear_regression.intercept_
```

```python
predictions = linear_regression.predict(rwd_scaled)
np.sqrt(mean_squared_error(wine_quality, predictions))
```
